## 🧠 ¿Qué es una API?

**API** significa *Application Programming Interface*.

Podemos entenderla como un **contrato de comunicación** entre sistemas:

> “Si me pides X de esta forma, yo te respondo Y en este formato.”


## 🔁 El ciclo Request → Response

Internet funciona bajo un modelo **Cliente – Servidor**.

```{mermaid}
sequenceDiagram
    participant Cliente as Cliente (Python / BI Tool)
    participant API as API Server
    participant DB as Database

    Cliente->>API: HTTP Request (GET /prices/AAPL)
    API->>DB: Query datos
    DB-->>API: Resultados
    API-->>Cliente: HTTP Response (200 + JSON)
```

### Flujo lógico

1. El cliente envía la petición  
2. El servidor valida y procesa  
3. Consulta fuentes internas  
4. Devuelve respuesta estructurada


## 🧩 Anatomía de un Request

Cuando consumes una API no envías solo una URL. Envías un mensaje estructurado.

```{mermaid}
flowchart LR
    A[Endpoint URL] --> B[HTTP Method]
    B --> C[Headers]
    C --> D[Body / Payload]
```

### 🔗 Endpoint (URL)

Representa el recurso:

- `https://api.github.com/users/google`
- `https://api.exchangerate.host/latest`

### 🧭 HTTP Methods

| Método | Acción | Uso en BI |
|---|---|---|
| GET | Leer | KPIs, precios, catálogos |
| POST | Crear | Transacciones, eventos |
| PUT / PATCH | Actualizar | Correcciones, cambios de estado |
| DELETE | Eliminar | Limpieza, bajas |

### 🧾 Headers

Metadatos típicos:

- `Authorization: Bearer <token>`
- `Content-Type: application/json`
- `Accept: application/json`

### 📦 Body (Payload)

Datos enviados al servidor (POST/PUT/PATCH):

```json
{
  "ticker": "AAPL",
  "cantidad": 10
}
```


## 📬 Anatomía del Response

```{mermaid}
flowchart LR
    A[Status Code] --> B[Headers]
    B --> C[Body JSON]
```

### 🚦 Status Codes (semáforo numérico)

| Código | Significado |
|---:|---|
| 200 | OK 🟢 |
| 201 | Created 🟢 |
| 400 | Bad Request 🔴 |
| 401 | Unauthorized 🔴 |
| 404 | Not Found 🟠 |
| 500 | Server Error 🔥 |

```{mermaid}
graph TD
    A[Request] --> B{Status Code}
    B -->|200| C[Procesar datos]

```


## 🌍 JSON: El Esperanto de los Datos

**JSON (JavaScript Object Notation)** es el formato estándar de intercambio de datos en APIs.

```json
{
  "ticker": "AAPL",
  "precio": 150.5,
  "activo": true,
  "historico": [140, 142, 145]
}
```

En Python, típicamente lo conviertes a diccionario con:

```python
data = response.json()
```


## 🧪 Práctica con `requests`

Aunque luego usaremos librerías especializadas (finanzas, economía, etc.), es clave entender la petición “cruda”.

### Instalación (con Poetry)

poetry add requests

> **Nota**: estos ejemplos requieren acceso a internet para ejecutarse.


````python
import requests

url = "https://api.github.com/zen"
response = requests.get(url, timeout=10)

print("Status:", response.status_code)
print("Body:", response.text)
````

💡 **Salida esperada (ejemplo)**

```text
Status: 200
Body: Keep it logically awesome.
```

La frase exacta puede cambiar; lo importante es observar:
- el `status_code`
- el tipo de cuerpo (`text/plain` en este endpoint)


````python
import requests

def buscar_usuario(usuario: str) -> None:
    url = f"https://api.github.com/users/{usuario}"
    response = requests.get(url, timeout=10)

    if response.status_code == 200:
        data = response.json()
        print("Name:", data.get("name"))
        print("Public repos:", data.get("public_repos"))
        print("Followers:", data.get("followers"))
    elif response.status_code == 404:
        print("Usuario no encontrado")
    else:
        print("Error inesperado:", response.status_code)

buscar_usuario("google")
````

💡 **Salida esperada (ejemplo)**

```text
Name: Google
Public repos: 200+
Followers: 100000+
```

Los valores cambian con el tiempo; lo importante es:
- parsear JSON con `.json()`
- usar `.get(...)` para evitar `KeyError`
- manejar códigos distintos a 200/404


## 🧠 Buenas prácticas mínimas (BI / Data)

- Define `timeout` en tus requests (evita cuelgues).
- Valida siempre `status_code` antes de usar el body.
- No asumas que el JSON trae siempre las mismas llaves.
- Separa “consumo de API” de “transformación analítica” (funciones distintas).
- Loggea errores con contexto (endpoint, params, status, snippet).
