# 1️⃣ ¿Qué es una API?

API significa:

**Application Programming Interface**

Una API es una forma **estandarizada de permitir que otros programas interactúen con tu sistema**.

Por ejemplo:

| Sistema | API |
|------|------|
| Google Maps | obtener coordenadas |
| Twitter | publicar tweets |
| Stripe | procesar pagos |
| Binance | consultar precios |
| OpenAI | generar texto |

---

## Ejemplo conceptual

Un cliente solicita información:

```
GET /price/AAPL
```

El servidor responde:

```json
{
  "ticker": "AAPL",
  "price": 185.2
}
```

La API define **cómo pedir información** y **qué respuesta recibir**.



# 2️⃣ Arquitectura Cliente-Servidor

Las APIs web siguen un modelo simple:

Cliente → Servidor → Respuesta

```mermaid
flowchart LR
    A[Cliente / Aplicación] --> B[HTTP Request]
    B --> C[Servidor FastAPI]
    C --> D[Procesamiento]
    D --> E[Respuesta JSON]
    E --> A
```



# 3️⃣ ¿Qué es FastAPI?

FastAPI es un framework moderno para construir APIs en Python.

Fue diseñado para:

- alto rendimiento
- tipado fuerte
- validación automática
- documentación automática
- desarrollo rápido

---

## Ventajas principales

### 🚀 Velocidad

FastAPI usa **ASGI**, lo que permite manejar muchas solicitudes concurrentes.

### 🧠 Tipado

Utiliza **type hints de Python** para validar datos automáticamente.

### 📚 Documentación automática

FastAPI genera automáticamente:

- Swagger UI
- ReDoc

### 🧩 Integración con Pydantic

Permite validar entradas y salidas usando modelos estructurados.



# 4️⃣ Instalación

Para usar FastAPI necesitamos instalar:

Con Poetry:


# 5️⃣ Nuestro primer API

Un servidor FastAPI mínimo se ve así:


````python

from fastapi import FastAPI

app = FastAPI()


@app.get("/")
def root():
    return {"message": "SmartPortfolio API is running"}
````


Este código define:

- una aplicación FastAPI
- un endpoint `/`
- una función que devuelve JSON



# 🧩 Del Notebook al Servidor Real

El notebook sirve **para aprender conceptos**, pero las APIs reales se ejecutan desde archivos Python.

Por eso debemos mover el código a un archivo dentro del proyecto.

---

## Estructura recomendada

Dentro del repositorio:

```
smart-portfolio-core/
│
├── src/
│   └── smart_portfolio_api/
│       └── main.py
```

---

## Contenido de `main.py`


````python

from fastapi import FastAPI

app = FastAPI(
    title="SmartPortfolio API",
    version="0.1.0"
)


@app.get("/")
def root():
    return {"message": "SmartPortfolio API is running"}


@app.get("/health")
def health():
    return {"status": "ok"}


@app.get("/version")
def version():
    return {
        "service": "smartportfolio",
        "version": "0.1.0"
    }
````


# 6️⃣ Ejecutar el servidor

Desde la raíz del proyecto ejecuta:

```bash
uvicorn src.smart_portfolio_api.main:app --reload
```

Explicación del comando:

```
src.smart_portfolio_api.main
```
→ ruta al archivo `main.py`

```
:app
```
→ variable que contiene la aplicación FastAPI

La opción `--reload` reinicia automáticamente el servidor cuando cambias el código.



# 7️⃣ Probar la API

# 8️⃣ Documentación automática
